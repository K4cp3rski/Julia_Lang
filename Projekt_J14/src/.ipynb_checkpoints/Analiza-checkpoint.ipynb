```julia
using DataFrames
using CSV
using GRUtils
using StatsBase
using HypothesisTests
using Printf
using LsqFit
using LinearAlgebra
using Distributions
using Roots
using Dierckx
```

```julia
filenames = ["0041.Spe", "0485.Spe", "0994.Spe", "1499.Spe", "2001.Spe", "2507.Spe", "3001.Spe", "3250.Spe", "3524.Spe", "3754.Spe", "3850.Spe", "3898.Spe", "3949.Spe", "4000.Spe", "4062.Spe"]

files = ["0041", "0485", "0994", "1499", "2001", "2507", "3001", "3250", "3524", "3754", "3850", "3898", "3949", "4000", "4062"]
x = 1:1:4096
```

```julia
df = DataFrame(CSV.File(filenames[1], skipto=13, header=false, footerskip=16, delim="\t"))
for file in filenames
    if file == filenames[1]
        continue
    else
        df_new = DataFrame(CSV.File(file, skipto=13, header=false, footerskip=15, delim="\t"))
        df = hcat(df, df_new, makeunique=true)
    end
end
rename!(df, Symbol.(files))
```

## Kalibracja

#### Kalibracja ciśnienia

```julia
@. lin(x, p) = p[1] * x + p[2]
```

```julia
ps_v = [46.4, 0.41]
ps_bar = [1.0049, 0.004]

fps_real = curve_fit(lin, ps_v, ps_bar, [0.218,-0.0049])
fps_fun(x) = (x->lin(x, fps_real.param))(x)

ps_pts = [1.023,0]
len_pts = [39.5, 0]

flen_fun = curve_fit(lin, ps_pts, len_pts, [0.218,-0.0049])
flen_real(x) = (x->lin(x, flen_fun.param))(x)

f_v_to_len(x) = flen_real(fps_fun(x))./10
```

```julia
f_v_to_len(40.62)
```

#### Kalibracja Energii

```julia
kal_names = ["2123", "1623", "1123", "623", "123"]
df_kal = DataFrame(CSV.File(filenames[1], skipto=13, header=false, footerskip=16, delim="\t"))
for file in kal_names
    if file == kal_names[1]
        continue
    else
        df_new = DataFrame(CSV.File(("kalibracja_"*file*".Spe"), skipto=13, header=false, footerskip=15, delim="\t"))
        df_kal = hcat(df_kal, df_new, makeunique=true)
    end
end
rename!(df_kal, Symbol.(kal_names))
```

```julia
peaks = []
energies_kal = []
base_energy = 5276
mult = base_energy/parse(Int64, kal_names[1])
for pulser_channel in kal_names
    push!(energies_kal, parse(Int64, pulser_channel)*mult)
end
for col in eachcol(df_kal)
    rownum = findfirst(==(maximum(col)), col) 
    push!(peaks, rownum)
    println(rownum)
end
```

```julia
energies_kal
```

```julia
fcal_peaks = curve_fit(lin, peaks[1:end-1], energies_kal[1:end-1], [-20,1.0])
fcal(x) = (x->lin(x, fcal_peaks.param)/1000)(x)
```

```julia
hold(false)
fig = Figure((1000, 600), "pix")
for col in eachcol(df_kal)
    plot(fcal.(x), col)
    hold(true)
end
legend("Kanał Pulsera 2123", "Kanał Pulsera 1623", "Kanał Pulsera 1123", "Kanał Pulsera 623", "Kanał Pulsera 123", location=2)
xticks(0.25, 2)
xlabel("Energy [MeV]")
ylabel("Counts")
display(gcf())
savefig("peaks_kal_energy.pdf", gcf())
```

```julia
res = Array{Float64, 1}()
press = Array{Float64, 1}()
for (col, pressure) in zip(eachcol(df), names(df))
    append!(res, float(sum(col)))
    append!(press, parse(Float64, pressure)/100)
end
```

```julia
press
```

```julia
fn = curve_fit(lin, press[end-5:end-2], res[end-5:end-2], [-20,1.0])
xf = 37.75:0.1:41
x_tot = 0:0.001:press[end-1]
x_gauss = 0:0.001:60
x_der = 38:0.001:40
f_lin(p) = fn.param[1]*p + fn.param[2]
```

```julia
spl = Spline1D(press, res, k=3, bc="extrapolate", s=10)
```

```julia
function RK4(func, X0, t, alpha,  beta, delta, gamma)
    """
    Runge Kutta 4 solver.
    """
    dt = t[1] - t[0]
    nt = len(t)
    X  = np.zeros([nt, len(X0)])
    X[0] = X0
    for i in range(nt-1)
        k1 = func(X[i], t[i], alpha,  beta, delta, gamma)
        k2 = func(X[i] + dt/2. * k1, t[i] + dt/2., alpha,  beta, delta, gamma)
        k3 = func(X[i] + dt/2. * k2, t[i] + dt/2., alpha,  beta, delta, gamma)
        k4 = func(X[i] + dt    * k3, t[i] + dt, alpha,  beta, delta, gamma)
        X[i+1] = X[i] + dt / 6. * (k1 + 2. * k2 + 2. * k3 + k4)
    end
    return X
end
```

```julia
function derivative(X, t, p)
    E = X
    x = t
    κ = 0.307
    @. dotE = - κ()/() log()
    return [dotE]
end
```
